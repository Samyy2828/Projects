# **This is a function to split the amount the person has into 3 parts**

```python
def split_num(n):
  parts = 3
  result = []
  for i in range(parts-1):
    x = np.random.randint(1,n-parts+i+1)
    n = n - x
    result.append(x)
  result.append(n)
  return result
```

# **This is a function to deduct**

This function would select the top 3 allocations and subtract it from the player, then bid would give the top three with others being zero

```python
def b(deleted,arr_player):
  before = deleted
  final = []
  allocation = 3
  deleted = deleted.tolist()
  deleted.sort(reverse=True)
  deleted = deleted[:3]
  for i in range(len(before)):
    if before[i] in deleted:
      final.append(before[i])
    else:
      final.append(0)
  b = final
  arr_player = arr_player-final
  return [arr_player,b]
```

# **This is a function to check if all players have been allocated 3 stocks**

```python
def checker(d):
  c = 0
  for i,v in d.items():
    actual_length = len(d)
    if len(v)==5:
      c+=1
  if c==actual_length:
    return 0
  else:
    return 1
```

# **Function to get arr_not from dictionary**

```python
def get_not(d,player_name):
  val = d[player_name]
  if len(val)==0:
    return []
  elif len(val)==1:
    return list(map(int, val))
  else:
    val = val.split(',')
    arr_not = list(map(int, val))
    return arr_not
```

# **Function to insert bid to d**

```python
def inserting_to_d(d,bid,stock_name):
  bid_dict = bid.to_dict()
  for i,v in bid_dict.items():
    if v>0:
      allo = str(stock_name)
      if len(d[i])<1:
        d[i] = allo
      else:
        d[i] = d[i]+','+allo
  return d
```

# **Delete players who have been allocated more than 3 stocks**

```python
#Empty the pockets of players who have got 3 stocks
def delete_player(d,arr_player):
  for i,v in d.items():
    if len(v)==5:
      arr_player[i] = 0.0
  return arr_player
```
